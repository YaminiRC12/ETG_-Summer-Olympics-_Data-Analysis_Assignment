**1. In how many cities Summer Olympics is held so far?**

```python
print('In',len(df['City'].unique()),"Cities Summer Olympics is held so far")
```


**2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
goldmedal_data=df[df['Medal']=="Gold"]
goldmedal_data.head()
```

```python
goldmedal_data['Sport'].unique() #display all unique sports
data=[] 
for sport in goldmedal_data['Sport'].unique():
  data.append([sport,len(goldmedal_data[goldmedal_data['Sport']==sport])])
data =pd.DataFrame(data,columns=['Sport','gold medal']) #to make dataframe we use DataFrame 
data=data.sort_values(by ='gold medal',ascending = False).head() #to sort the values i.e freq in descending manner
data.plot(x='Sport',y='gold medal',kind ='bar',figsize =(5,5)) #plot graph 
data.head()
```

**3. Which sport is having most number of medals so far? (Top 5)**

```python
data=[]
for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])
data=pd.DataFrame(data,columns=['Sport','Medals'])
data=data.sort_values(by='Medals',ascending=False).head(5)
data.plot(x='Sport',y='Medals',kind='bar',figsize=(5,5))
data.head()
```

**4. Which player has won most number of medals? (Top 5)**

```python
data=[]
for athlete in df['Athlete'].unique():
    data.append([athlete,len(df[df['Athlete']==athlete])])
data=pd.DataFrame(data,columns=['Athlete','Medals'])
data=data.sort_values(by='Medals',ascending=False).head()
data.plot(x='Athlete',y='Medals',kind='bar',figsize=(5,5))
data.head()
```

**5. Which player has won most number Gold Medals of medals? (Top 5)**

```python
data=[]
df=df[df['Medal']=='Gold']
for athlete in df['Athlete'].unique():
    data.append([athlete,len(df[df['Athlete']==athlete])])
data=pd.DataFrame(data,columns=['Athlete','Gold Medals'])
data=data.sort_values(by='Gold Medals',ascending=False).head(5)
data.plot(x='Athlete',y='Gold Medals',kind='bar',figsize=(5,5))
data.head()
```

**7. Which event is most popular in terms on number of players? (Top 5)**

```python
data=[]
for players in df['Event'].unique():
    data.append([players,len(df[df['Event']==players])])
data=pd.DataFrame(data,columns=['Event','No.of.Players'])
data=data.sort_values(by='No.of.Players',ascending=False).head()
data.plot(x='Event',y='No.of.Players',kind='bar',figsize=(5,5))
data.head()
```


**8. Which sport is having most female Gold Medalists? (Top 5)**


```python
data=[]
df=df[df['Gender']=='Women']
df=df[df['Medal']=='Gold']
for sport in df['Sport'].unique():
    data.append([sport,len(df[df['Sport']==sport])])
data=pd.DataFrame(data,columns=['Sport','Female Player'])
data=data.sort_values(by='Female Player',ascending=False).head()
data.plot(x='Sport',y='Female Player',kind='bar',figsize=(5,5))
data.head(5)
```
